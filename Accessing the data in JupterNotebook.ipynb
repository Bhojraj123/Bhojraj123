#### Accessing Data from SQLite DB

```python
import sqlite3 as sql

# Connect to the database
conn = sql.connect('demo.db')

# Create a cursor object
cur = conn.cursor()

# Check if the table exists, and create it if it doesn't
cur.execute("""
CREATE TABLE IF NOT EXISTS movie(
    title TEXT,
    year INTEGER,
    score REAL
)
""")

# Fetch all table names from sqlite_master
res = cur.execute("SELECT name FROM sqlite_master")
print("All tables in the database:", res.fetchall())  # This will print all the tables in the database

# Check if the table 'spam' exists
res = cur.execute("SELECT name FROM sqlite_master WHERE name='spam'")
print("Does 'spam' table exist?", res.fetchone() is None)  # This will print True if the table 'spam' does not exist

# Insert data into the movie table
cur.execute("""
    INSERT INTO movie VALUES
        ('Monty Python and the Holy Grail', 1975, 8.2),
        ('And Now for Something Completely Different', 1971, 7.5)
""")

# Commit the changes and close the connection
conn.commit()

```

```python
# query data from database
query = "SELECT * FROM movie"
results = cur.execute(query).fetchall()
res = cur.execute("SELECT score FROM movie")
print(res.fetchall())
data = [
    ("Monty Python Live at the Hollywood Bowl", 1982, 7.9),
    ("Monty Python's The Meaning of Life", 1983, 7.5),
    ("Monty Python's Life of Brian", 1979, 8.0),
]
cur.executemany("INSERT INTO movie VALUES(?, ?, ?)", data)
conn.commit()
```
