#### Web Scrapping Using python

```python
# Importing the necessary modules (pandas, numpy, matplotlib.pyplot, seaborn)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# To perform web scraping, you should also import the libraries shown below. 
# The urllib.request module is used to open URLs. 
# The Beautiful Soup package is used to extract data from html files. 
# The Beautiful Soup library's name is bs4 which stands for Beautiful Soup, version 4.

from urllib.request import urlopen
from bs4 import BeautifulSoup
# Getting the html of the page is just the first step.
# After importing necessary modules,specify the URL containing the dataset and pass it to urlopen() to get the html of the page.
url = "http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
html
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
soup.find_all('a')
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# To print out table rows only, pass the 'tr' argument in soup.find_all().

# Print the first 10 rows for sanity check
rows = soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

#### Data Manipulation and Cleaning

```python
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
col_labels = soup.find_all('th')
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df5.info()
df5.shape
```

```python
df6 = df5.dropna(axis=0, how='any')
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
df7.rename(columns={'[Place': 'Place'},inplace=True)
df7.rename(columns={' Team]': 'Team'},inplace=True)
df7.head()
```

```python
df7['Team'] = df7['Team'].str.strip(']')
df7.head()
```

#### Data analysis and Visualization

```python
print(df7.columns.tolist())
```

```python
print(df7.columns.tolist())
```

```python
# Strip spaces from the column names
df7.columns = df7.columns.str.strip()

# Now check the columns again
print(df7.columns.tolist())
```

```python
time_list = df7['Time'].tolist()
print(time_list)
```

```python
time_mins = []
for i in time_list:
    parts = i.split(':')
    
    if len(parts) == 3:  # Case with hours, minutes, and seconds
        h, m, s = parts
        math = (int(h) * 3600 + int(m) * 60 + int(s)) / 60
    elif len(parts) == 2:  # Case with only minutes and seconds
        m, s = parts
        math = (int(m) * 60 + int(s)) / 60
    else:
        # Handle unexpected cases, you could skip or raise an error
        continue
    
    time_mins.append(math)

print(time_mins)
```

```python
df7['Runner_mins'] = time_mins
df7.head()
```

```python
df7.describe(include=[np.number])
```
